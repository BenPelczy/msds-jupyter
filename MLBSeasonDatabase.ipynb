```python
import numpy as np
import pandas as pd
import requests
import json
```

```python
url = "https://api.sportradar.com/mlb/trial/v7/en/league/teams.json?api_key=MoxxuC14537WIDKvXqnAHTKHiJDgHtpK4PiwZZmN"

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

teams = pd.json_normalize(json.loads(response.text), record_path = ["teams"])
teams.head()
```

```python
stats = pd.DataFrame()
for i in teams['id']:
    url = "https://api.sportradar.com/mlb/trial/v7/en/seasons/2023/REG/teams/"+i+"/statistics.json?api_key=MoxxuC14537WIDKvXqnAHTKHiJDgHtpK4PiwZZmN"
    headers = {"accept": "application/json"}
    response = requests.get(url, headers=headers)
    stats = pd.concat([stats, pd.json_normalize(json.loads(response.text))])
stats.head()
```

```python
players = pd.DataFrame()
for i in teams['id']:
    url = "https://api.sportradar.com/mlb/trial/v7/en/seasons/2023/REG/teams/"+i+"/statistics.json?api_key=MoxxuC14537WIDKvXqnAHTKHiJDgHtpK4PiwZZmN"
    headers = {"accept": "application/json"}
    response = requests.get(url, headers=headers)
    if 'players' in pd.json_normalize(json.loads(response.text)):
        players = pd.concat([players, pd.json_normalize(json.loads(response.text), record_path = 'players')])
players.head()
```

```python
stats.to_csv('MLBstats.csv')
players.to_csv('MLBplayers.csv')
```
